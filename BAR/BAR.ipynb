```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
pd.set_option('display.max_colwidth', None)

URL = "https://www.beyondallreason.info/"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
# print(soup.prettify())
# soup
```

```python
def get_class(link_UO):
    print(link_UO)
    UO = requests.get(link_UO)
    UO = BeautifulSoup(UO.content, "html.parser")
    df_return = []
    for UC in UO.find_all('a',attrs={'class':re.compile('unittypeselect w-inline-block')}):
        line = [link_UO]
        line.append(URL+UC.get('href'))
        line.append(UC.find('h1').text)
        line +=UC.find('div',attrs={'class':re.compile('label')}).text.split()
        line.append(UC.find('div',attrs={'class':re.compile('unit-description type')}).text)
        # print(UC.prettify())
        df_return.append(line)
        # break
    # df_return = pd.DataFrame(df_return)
    # display(df_return)
    return df_return

    
df_UO = []
for UO in soup.find_all('a',string=re.compile('Unit Overview',re.I)):
    UO=URL+UO.get('href')
    df_UO += get_class(UO)
    
    # break
df_UO = pd.DataFrame(df_UO)
display(df_UO)
```
